```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pytrends
from pytrends.request import TrendReq
```

```python
data = pd.read_csv('processed_search_trends_final.csv', index_col=1)
data
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
data
```

```python
len(data.index.values.tolist())
```

```python
def get_maxes():
    max_values = []
    max_countries = []
    for index, row in data.iterrows():
        #print(f"Max for {index}:")
        maximum = row.max()
        max_values.append(maximum)
        #print(maximum)
        ind = list(row).index(maximum)
        country = data.columns.values.tolist()[ind]
        max_countries.append(country)
        #print(country)
        dat = {"Search Terms":data.index.values.tolist(),
               "Max Diff":max_values,
               "Country":max_countries}
    return pd.DataFrame(dat)
def get_mins():
    min_values = []
    min_countries = []
    for index, row in data.iterrows():
        #print(f"Max for {index}:")
        minimum = row.min()
        min_values.append(minimum)
        #print(maximum)
        ind = list(row).index(minimum)
        country = data.columns.values.tolist()[ind]
        min_countries.append(country)
        #print(country)
        dat = {"Search Terms":data.index.values.tolist(),
               "Min Diff":min_values,
               "Country":min_countries}
    return pd.DataFrame(dat)
get_maxes()



```

```python
print(get_maxes().to_latex(index=False))
```

```python
get_mins()
```

```python
print(get_mins().to_latex(index=False))
```

```python
def get_avg_terms():
    avg_terms = data.loc[:,'Avg'].sort_values()
    indices = avg_terms.index.tolist()
    avg_terms = avg_terms.tolist()

    return pd.DataFrame({"Search Term":indices, "Average Diff":avg_terms})
get_avg_terms()
```

```python
print(get_avg_terms().to_latex(index=False))
```

```python
def get_avg_countries():
    avg_countries = data.loc['Avg'].sort_values()
    #print(avg_countries)
    indices = avg_countries.index.tolist()
    avg_countries = avg_countries.tolist()

    return pd.DataFrame({"Search Term":indices, "Average Diff":avg_countries})
get_avg_countries()
```

```python
print(get_avg_countries().to_latex(index=False))
```

```python
# Establish connection to API
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
def make_comparison_graph(term, c1, c2, time='today 12-m'):
    pytrends.build_payload([term], cat=0, timeframe = time, geo=c1, gprop='')
    df1 = pytrends.interest_over_time()
    pytrends.build_payload([term], cat=0, timeframe = time, geo=c2, gprop='')
    df2 = pytrends.interest_over_time()
    plt.plot(df1.index, df1.iloc[:,0], label=c1)
    plt.plot(df2.index, df2.iloc[:,0], label=c2)
    plt.title(f"{c1} vs. {c2}: Interest in {term}")
    plt.xlabel("Date")
    plt.ylabel("Interest")
    plt.legend()
    plt.savefig(f"{term}_{c1}_vs_{c2}.png")
    plt.show()
make_comparison_graph("Election Results", 'US',"MX")
make_comparison_graph("Boomer", 'US',"IN")
make_comparison_graph("Boomer", 'US',"CN")
make_comparison_graph("Coronavirus", 'US',"CN")
make_comparison_graph("Elon Musk", 'US',"DE")
```
